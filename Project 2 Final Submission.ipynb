## Scientific Question: How does the genetic sequence of the human APOE gene and the protein it encodes for, Apolipoprotein E, compare to the APOE gene and protein in mice, and if the ε4 allele, which is a single nucleotide polymorphism in ApoE, can increase risk of Alzheimer’s Disease in humans, could this same mutation have similar memory-loss effects in mice?

#### Alzheimer’s disease is a common disease that causes dementia in the elderly and is characterized by decline in memory and other cognitive functions. One known genetic risk factor for Alzheimer’s disease is the ε4 allele of Alipoprotein E (ApoE) (Kim, 2009). APOE gene is a protein-encoding gene located in humans on Chromosome 19 and encodes for Apolipoprotein E (Bekris, 2011). The ε4 allele is a single nucleotide polymorphism in exon 4 of ApoE, where at position 130 a C is switched for a R, and this mutation is a genetic risk factor for Alzheimer’s disease (Uniprot).

#### While the pathway for how the ε4 allele increases Alzheimer’s risk is not fully understood, there is evidence to support multiple potential mechanisms. One suggests that ApoE isoforms have isoform-specific binding to amyloid beta peptide, which is involved in the pathogenesis of Alzheimer’s disease (Strittmatter, 1993). Another suggests that the ε4 allele affects amyloid beta aggregation and clearance in the brain, resulting in buildup of amyloid plaques in the brain (Kim, 2009). While the exact mechanism is unknown, there is a clear correlation between the ε4 ApoE allele and increased risk for Alzheimer’s disease.

#### Dementia and other cognitive function disorders are not unique to humans, and mouse models are often used to study Alzheimer’s disease due to the conservation of APOE gene in mice (Maloney, 2007). The APOE gene is also conserved in the chimpanzee (Pan troglodytes), dog (Canis lupus familiaris), cow (Bos taurus), mouse (Mus musculus), rat (Rattus norvegicus), and frog (Xenopus tropicalis) (NCBI). This suggests that mutations that occur in APOE in humans could also occur in these species.

#### NCBI (https://www.ncbi.nlm.nih.gov/gene/348) and UniProt (https://www.uniprot.org/uniprot/P02649) databases contain detailed information about APOE and ApoE including gene and protein sequence, isoforms, specific mutations that occur and where these occur, and across which species the gene and protein are conserved. The APOE and ApoE sequences used for analysis were downloaded from NCBI and UniProt.

## Scientific Hypothesis: If the mutation in ApoE that causes the ε4 allele, the risk factor for Alzheimer’s disease in humans, were to occur in mice and other animals and result in similar memory-loss effects, then the protein encoded by APOE gene, Apolipoprotein E, should be highly similar in the part of the sequence around where the snp occurs.

#### The snp that causes the ε4 allele in ApoE, the risk factor for Alzheimer’s disease, is located at position 130 in exon 4 of ApoE. To compare the effects of this mutation in humans versus mice, the sequence of APOE gene and Apolipoprotein E were compared in both humans, mice, and other animals, around where the snp occurs in ApoE.

#### To first determine if the APOE gene in humans and mice was similar enough to compare for mutations and function, a pairwise sequence analysis was performed to test for similarity between the two sequences. The alignment was a local pairwise alignment, performed between exon 4 of human APOE versus exon 4 of mouse APOE, rather than the entire gene sequence, since the region around the snp was the region of interest. The APOE sequence similarity in exon 4 between humans and mice was visualized by a dot plot. Sequence similarity was shown as black diagonal line across the plot. Sequence similarity is important because it determines that the two gene sequences are similar enough to be studied comparatively and produce significant results.

#### APOE gene is conserved in the chimpanzee, dog, cow, mouse, rat, and frog, and APOE encodes Apolipoprotein E in each species. If the protein isoforms encoded by APOE in each of these species are similar, this should support the hypothesis that the APOE gene has an evolutionarily conserved function across species, and therefore a mutation in ApoE should have a similar sequence effect in the ApoE isoforms across species. To analyze this, a local multiple sequence alignment was performed for part of the ApoE isoform sequences for the species listed above. The alignment was conducted using the region of ApoE corresponding to where the snp occurs in human ApoE, since this was the region of interest.

#### The multiple sequence alignment was visualized by a sequence logo. Conserved nucleotides show evolutionarily conserved sequence similarity, and if the sequences were to have conserved nucleotides in the same positions across multiple species, this would provide evidence of evolutionarily conserved sequence of ApoE. This would support the hypothesis that across species, there would be shared similar sequence effects if the mutation that causes the ε4 allele in human ApoE were to occur across species.

### Part 2: Load in the data and perform Bioinformatics Analyses (Pairwise Sequence Alignment and Multiple Sequence Alignment)
#### 2A. Pairwise Sequence Alignment Between Exon 4 of Human APOE Gene vs Exon 4 of Mouse APOE Gene
#### For the local pairwise sequence alignment, the DNA sequences of human and mouse APOE were downloaded from NCBI and cut to only include the sequences for exon 4, the region where the snp occurs in ApoE. These two sequences were stored in separate files, ‘APOE_homo_sapiens’ (human) and ‘APOE_mus_musculus’ (mouse), both of which are .txt files. For conducting the pairwise sequence alignment, Bio.SeqIO.read was used to read the file types in fasta format, which is a better text-based format for representing nucleotide sequences. Additionally, the .seq function was used on these files to read the contents as a nucleotide sequence, rather than just letters, in order to perform the pairwise sequence alignment. The pairwise sequence alignment takes two sequences, looks for nucleotide mismatches and places where insertions and deletions could have occurred, and re-formats the DNA sequences to find the best similarity possible. Conserved nucleotides are scored with a +1 and non-conserved nucleotides are scored with a -1. A high similarity score means increased chance that the two sequences are of similar evolutionary origin. 

```python
# Define a global variable to read file 'APOE_homo_sapiens_exon4' as a fasta file using bio function SeqIO.read
APOE_homo_sapiens = SeqIO.read("APOE_homo_sapiens_exon4.txt", "fasta")
# Define a global variable where contents of the APOE_homo_sapiens file are in sequence form using bio function .seq
APOE_homo_sapiens_seq = APOE_homo_sapiens.seq
# Print sequence 'APOE_homo_sapiens_seq' to confirm correct sequence; not shown in final output
# print(APOE_homo_sapiens_seq)

# Define a global variable to read file 'APOE_mus_musculus_exon4' as a fasta file using bio function SeqIO.read
APOE_mus_musculus = SeqIO.read("APOE_mus_musculus_exon4.txt", "fasta")
# Define a global variable where contents of the APOE_mus_musculus file are in sequence form using bio function .seq
APOE_mus_musculus_seq = APOE_mus_musculus.seq
# Print sequence 'APOE_mus_musculus_seq' to confirm correct sequence; not shown in final output
# print(APOE_mus_musculus_seq)

# Conduct pairwise sequence alignment between sequences, 'APOE_homo_sapiens_seq' and 'APOE_mus_musculus_seq'
alignments = pairwise2.align.globalxx(APOE_homo_sapiens_seq, APOE_mus_musculus_seq)

# Print the results of the pairwise sequence alignment for all alignments conducted
for alignment in alignments:
    print(format_alignment(*alignment))
    # Breaking to only display first result for final output
    break
```

#### 2A. Multiple Sequence Alignment Between snp Region of ApoE Isoforms for Human, chimpanzee, Dog, Cow, Mouse, Rat, and Frog 
#### For the local multiple sequence alignment, the protein isoforms of ApoE, encoded for by APOE, for each of the species being compared (human, chimpanzee, dog, cow, mouse, rat, and frog) were downloaded from NCBI. The sequences were cut to only include a 20-nucleotide region, approximately positions 120-140, since the snp in human ApoE occurs at position 130. All 7 of these 20-nucleotide long sequences were stored in a single file, ‘ApolipoproteinE_seq.fasta,’ which was already in .fasta format, with each sequence separated by a new line and a header listing the name of the species. The multiple sequence alignment takes all 7 sequences, of the same length, and aligns them without indels or spacers to see the nucleotides that are conserved and not conserved in each position. More nucleotides conserved in the proteins is a better indication of evolutionarily conserved sequence of ApoE and function of the APOE gene.

```python
# Function that parses the contents of a fasta file containing 1 or more sequences
# Returns a list, where each element is the string contents of a sequence
def getSeqsFromFile(file_contents):
    # Each sequence starts with >, so split contents in file by > to seperate sequences
    seq_delimeter = ">"
    raw_seqs = file_contents.split(seq_delimeter)
    
    # First element in list after splitting will be empty string, so skip over that element
    raw_seqs = raw_seqs[1:]

    # Initialize return list that will store the parsed sequence strings
    seqs = []
    # For each sequence in the file
    for raw_seq in raw_seqs:
        # Find the first newline character, which marks the end of the sequence header
        first_newline = raw_seq.find("\n")
        # Sequence contents starts after the end of the header
        seq = raw_seq[first_newline+1:]
        # Trim trailing whitespace
        seq = seq.strip()
        # Append parsed sequence contents to return list
        seqs.append(seq)

    # Return the list containing the string contents of each sequence 
    return seqs
```

```python
# Open the fasta file containing the snp region of the protein sequences for each of the species
file = open('ApolipoproteinE_snp.fasta')
# Read the contents of the file into a global variable, file_contents
file_contents = file.read()
# Close the opened fasta file
file.close()

# Parse the fasta file contents using the getSeqsFromFile() function and store results in global variable, seqs
seqs = getSeqsFromFile(file_contents)
```

### Part 3A: Create a dot plot to show the sequence similarity from pairwise sequence alignment for exon 4 of APOE gene in humans and mice.
#### A dot plot is a plotting method that shows sequence similarity between two sequences, where one sequence is on the x-axis and the other is on the y-axis. Here, the sequence for exon 4 of human APOE is expressed on the x-axis, and the sequence for exon 4 of mouse APOE is expressed on the y-axis. Everywhere the sequences are similar is marked in black, and where the sequences are dissimilar is marked in white. This dot plot uses window size 3 and allows for no mismatches. For the dot plot, if the sequences are similar, a black diagonal line appears clearly across the dot plot.

```python
# Define a global variable, window, that determines the window size for doing
# individual nucleotide comparison in the pairwise sequence alignment
window = 3
# Get the string contents of both sequences
seq_one = str(APOE_homo_sapiens_seq)
seq_two = str(APOE_mus_musculus_seq)

# Define a global variable, data, that will store the pairwise sequence alignment comparisons
# at each nucleotide, comparing within the defined window size
data = [
    [
        (seq_one[i : i + window] != seq_two[j: j + window])
        for j in range(len(seq_one) - window)
    ]
    for i in range(len(seq_two) - window)
]
```

```python
# Use the pylab function figure() to increase the dot plot figure size
pylab.figure(figsize=(8, 8))
# Use the pylab function gray() to make the dot plot color scheme gray
pylab.gray()
# Use the pylab function imshow() to pass in the pairwise sequence alignment data results to display
pylab.imshow(data)
# Set the x axis label for the dot plot
pylab.xlabel("APOE_homo_sapiens (length %i bp)" % (len(APOE_homo_sapiens_seq)))
# Set the y axis label for the dot plot
pylab.ylabel("APOE_mus_musculus (length %i bp)" % (len(APOE_mus_musculus_seq)))
# Set the title of the dot plot
pylab.title("Dot plot using window size %i\n(allowing no mis-matches)" % window)
# Display the dot plot using the pylab function show()
pylab.show()
```

### Part 3B: Create a sequence logo to show the conserved nucleotides from multiple sequence alignment of the snp-region of the proteins encoded for by APOE in humans, chimpanzee, dog, cow, mouse, rat, and frog. 
#### A sequence logo is a visualization tool that is used here to visualize the sequence similarity from the multiple sequence alignment. Sequence logos use a probability matrix to compare the nucleotides present in each location, across all sequences being aligned. Sequence positions 1-20 are indicated along the x-axis, and the probability of each nucleotide found is on the y-axis. For each position in the sequence, the nucleotides found in that position are displayed on the graph, with the nucleotide found in the highest frequency enlarged to show it is highly conserved across sequences. This is done for each position to show which nucleotides are found in each position, and the frequency of these nucleotides, ultimately showing the levels of nucleotide conservation across sequences. Higher frequency of conserved nucleotides supports the hypothesis of similar evolutionary origin of the different sequences.

```python
# Function that takes in a list of sequences (of the same length) and produces a probability matrix for the nucleotides
# at each position
# Returns a probability matrix in the form of a pandas dataframe, where row indices represent sequence positions
# (0, 1, 2, ...) and columns represent the various possible nucleotides (R, G, A, ...)
def createProbabilityMatrix(seqs):
    # List of nucleotides seen across all sequences, aka the alphabet
    nucleotides = list(set(list("".join(seqs))))
    
    # Possible sequence positions, assuming all sequences are of same length
    positions = list(range(len(seqs[0])))
    
    # Initialize dataframe to contain probabilities, initially all set to 0
    df = pd.DataFrame(0.0, index=positions, columns=nucleotides)

    # For each sequence position
    for pos in range(len(seqs[0])):
        # Initialize a dictionary to track the counts of each nucleotide at this position
        counts = {}
        # For each sequence
        for seq in seqs:
            # Determine the nucleotide at this position for this sequence 
            nucleotide = seq[pos]
            # Increment the count for this nucleotide
            if nucleotide in counts:
                counts[nucleotide] += 1
            else:
                counts[nucleotide] = 1
        
        # For all nucleotides seen at this position
        for nucleotide in counts:
            # Calculate the probability of this nucleotide
            prob = counts[nucleotide] / len(seqs)
            # Update the probability matrix, at this position and nucleotide
            df.iloc[pos][nucleotide] = prob

    # Return the probability matrix
    return df
```

```python
# Calculate the probability matrix using the createProbabilityMatrix() function
# and store the results in the global variable, prob_matrix
prob_matrix = createProbabilityMatrix(seqs)

# Create Logo object using sequence logo function .Logo
# Use the data from 'prob_matrix' to create the sequence logo
# Specify other display options, inluding the width, how the nucleotides are displayed, and the colorscheme
logo = logomaker.Logo(prob_matrix,
                         width=.8,
                         vpad=.05,
                         fade_probabilities=True,
                         stack_order='small_on_top',
                         color_scheme='dodgerblue'
                     )

# Style using Logo methods
logo.style_spines(spines=['left', 'right'], visible=False)

# Style the sequence logo using axes methods
# Set the labaled x-axis ticks as the positions 0-19 of the sequences using the sequence logo function .ax.set_xticks
logo.ax.set_xticks(range(len(prob_matrix)))
# Set the labaled y-axis ticks as the probabilities 0-1 counting by 0.25 using the sequence logo function .ax.set_yticks
logo.ax.set_yticks([0, 0.25, 0.5, 0.75, 1])
logo.ax.axvline(2.5, color='k', linewidth=1, linestyle=':')
# Label the x-axis as 'Sequence Positon' using the sequence logo function .ax.set_xlabel
logo.ax.set_xlabel('Sequence Position')
# Label the y-axis as 'Probability' using the sequence logo function .ax.set_ylabel
logo.ax.set_ylabel('Probability')
```
